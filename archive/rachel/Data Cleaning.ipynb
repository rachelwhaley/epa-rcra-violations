```python
evals_counts = evals_df.groupby("ID_NUMBER").size().reset_index().rename(columns={0: 'Count'}).sort_values("Count",ascending=False)# .head(20)
```

```python
facilities_w_counts = pd.merge(evals_counts, facilities_df, left_on='ID_NUMBER', right_on='ID_NUMBER', how='left')
facilities_w_counts.head(20)
```

```python
# need to separately count the found violations and not found violations and do a percentage
# need to separate out the FULL_ENFORCEMENT column
# need column with date of first eval and date of most recent eval
# need column with # of months since last violation found
```

```python
# create violations column with a 1 for yes and a 0 for no or unknown

evals_df["FOUND_VIOLATION_01"] = pd.Series(np.where(evals_df.FOUND_VIOLATION.str.contains("Y"), 1, 0), evals_df.index)
evals_df.head()
```

```python
violation_sums = evals_df.groupby("ID_NUMBER")['FOUND_VIOLATION_01'].sum().reset_index().rename(columns={'FOUND_VIOLATION_01': 'Sum_Violations'}).sort_values("Sum_Violations",ascending=False)
violation_sums.head(20)
```

```python
facilities_w_violations = pd.merge(facilities_w_counts, violation_sums, left_on='ID_NUMBER', right_on='ID_NUMBER', how='left')
facilities_w_violations.head(20)
```

```python
facilities_w_violations.plot(x='Count', y='Sum_Violations', style='o')
```

```python
facilities_w_violations.describe()
```

```python
max_date = evals_df.groupby("ID_NUMBER").agg({'EVALUATION_START_DATE':'max'})[['EVALUATION_START_DATE']].reset_index().rename(columns={'EVALUATION_START_DATE': 'MostRecentEval'})

# .max(evals_df["EVALUATION_START_DATE"]).reset_index()# .rename(columns={0: 'MostRecentEval'}).sort_values("MostRecentEval",ascending=False).head(20)
# max_date.head()

facilities_w_violations = pd.merge(facilities_w_violations, max_date, left_on='ID_NUMBER', right_on='ID_NUMBER', how='left')
facilities_w_violations.head()
```

```python
from datetime import datetime
facilities_w_violations["MostRecentEval_x"] = pd.to_datetime(facilities_w_violations["MostRecentEval_x"], format='%m/%d/%Y', errors='coerce')

                                        
```

```python
facilities_w_violations["NumMonthsSinceEval"] = (datetime.today() - facilities_w_violations["MostRecentEval_x"])/ (np.timedelta64(1, 'M'))
facilities_w_violations.head()
```
